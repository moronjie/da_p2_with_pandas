```python
df = df[['coaster_name', 'Location', 'Status', 'Manufacturer',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 'speed_mph', 'height_ft','Inversions_clean', 'Gforce_clean']].copy()

df
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
df
```

```python
df.columns
```

```python
df = df[~df.duplicated(subset=['coaster_name', 'Location','opening_date_clean'])].reset_index(drop=True)
df
```

```python
df['year_introduced'].value_counts().head(15).plot(kind='bar', figsize=(10, 5), title='Top 15 Years Introduced', xlabel='Year Introduced', ylabel='count')
plt.show()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.reset_index(drop=True, inplace=True)
df.head(10)
```

```python
df['speed_mph'].plot(kind='hist', bins=20, title='Speed Histogram')
```

```python
sns.kdeplot(df['speed_mph'], shade=True)
```

```python
sns.boxenplot(y='speed_mph', data=df)
```

```python
plt.scatter(data=df, x='speed_mph', y='height_ft')
plt.show()
```

```python
sns.lmplot(data=df, x='speed_mph', y='height_ft')
```

```python
df.head()
```

```python
df_corr = df[['speed_mph', 'height_ft', 'Inversions_clean', 'Gforce_clean']].corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True, cmap='coolwarm')
```

```python
# location with faster speed
df.groupby('Location')['speed_mph'].agg(['mean','count']).query('count > 1').sort_values('mean', ascending=False).plot(kind='bar', ylim=(0, 100))
```
